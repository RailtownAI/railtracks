```python
from typing import List, Any, Tuple
import requestcompletion as rc
import openai
import json


from enum import Enum

from dotenv import load_dotenv
load_dotenv()

```

```python

class WeatherForecaster(rc.nodes.Node[str]):
    
    @classmethod
    def system_prompt(cls) -> str:
        return """Your goal is to provide the user with a weather forecast. You should use the tool calls available to you to get the information you need."""
    
    @classmethod
    def tools(cls) -> list:
        
        return [{
    "type": "function",
    "function": {
        "name": "get_current_weather",
        "description": "Collects the current weather at the given location",
        "parameters": {
            "type": "object",
            "properties": {
                "location": {
                    "type": "string",
                    "description": "The location you would like the current weather. Just include the name of the city"
                }
            },
            "required": [
                "location"
            ],
            "additionalProperties": False
        },
        "strict": True
    }
},
            {
    "type": "function",
    "function": {
        "name": "get_weekly_forecast",
        "description": "Collects the weekly weather forecast at the given location",
        "parameters": {
            "type": "object",
            "properties": {
                "location": {
                    "type": "string",
                    "description": "The location you would like the forecast"
                }
            },
            "required": [
                "location"
            ],
            "additionalProperties": False
        },
        "strict": True
    },
            },

            {
    "type": "function",
    "function": {
        "name": "get_long_term_forecast",
        "description": "Collects the long term weather forecast at the given location",
        "parameters": {
            "type": "object",
            "properties": {
                "location": {
                    "type": "string",
                    "description": "The location you would like the forecast"
                }
            },
            "required": [
                "location"
            ],
            "additionalProperties": False
        },
        "strict": True
    },
            },
            ]
 
    
    def get_node_from_tool_call(self, tool_call_name: str):
        """
        Below you can see two ways one can have a node that is simply calling a function
        A: Use the provided `FunctionNode` class which only needs to be initilized with the
        required callable and keyword arguments
        B: Inheriting from `TerminalNode` and defining the required `invoke` method

        """
        # Using Option A    
        if tool_call_name == "get_current_weather":
            return lambda location: rc.nodes.library.FunctionNode(get_current_weather, location=location)
        
        # Using Option B - The defined node classes can be found at the bottom of the notebook
        if tool_call_name == "get_weekly_forecast":
            return WeeklyForecast
        if tool_call_name == "get_long_term_forecast":
            return LongTermForecast
    
    
    def __init__(self, mess_hist: list[dict]):
        super().__init__()
        self.message_history = mess_hist
        self.message_history.insert(0, {"role": "system", "content": f"{self.system_prompt()}"})
        
        

    def invoke(self) -> str:
        while True:
            try:
                client = openai.Client()
                response = client.chat.completions.create(messages=self.message_history, model="gpt-4o", tools=self.tools())
                
            except Exception as e:
                raise rc.exceptions.ResetException(self, str(e))
            
            
            if response.choices[0].message.tool_calls:
                for t_c in response.choices[0].message.tool_calls:
                    self.message_history.append({
        "role": "assistant",
        "function_call": {
            "name": f"{t_c.function.name}",
            "arguments": f"{t_c.function.arguments}"
        }
    })
                
                responses = self.call_nodes([rc.nodes.NodeFactory(self.get_node_from_tool_call(t_c.function.name), **json.loads(t_c.function.arguments)) for t_c in response.choices[0].message.tool_calls])
                
                handled_responses = [self.handle_response(r.data) for r in responses]
                
                for request_id, tool_resp in zip([x.function.name for x in response.choices[0].message.tool_calls], handled_responses):
                    self.message_history.append({"role": "function", "name": request_id, "content": tool_resp})
            
            if response.choices[0].message.content is not None:
                break
        
        return response.choices[0].message.content
                
    def handle_response(self, output: Any) -> str:
        if isinstance(output, Tuple):
            return str(output)

        if isinstance(output, str):
            return output
        
        if isinstance(output, list):
            return "\n".join(str(item) for item in output)
        
        raise rc.exceptions.FatalException(self, f"Unknown output {output} {output}")
    
        
    @classmethod
    def pretty_name(cls):
        return "Top Level Weather"
```

```python
from typing import Callable
        
        
class WeeklyForecast(rc.nodes.Node[List[str]]):
    
    def __init__(self, location: str):
        super().__init__()
        self.location = location
        
    def invoke(
        self,
    ) -> List[str]:
        try:
            forecast = get_weekly_forecast(self.location)
            return [f"{temp} degrees and {w_type.value}" for temp, w_type in forecast]
        except LocationNotFoundException as e:
            raise rc.exceptions.CompletionException(self, str(e), [])
        
    @classmethod
    def pretty_name(cls):
        return "Weekly Forecast"
        
class LongTermForecast(rc.nodes.Node[str]):
    
    def __init__(self, location: str):
        super().__init__()
        self.location = location
        
    def invoke(
        self,
    ) -> str:
        try:
            forecast = get_long_term_forecast(self.location)
            return forecast
        except LocationNotFoundException as e:
            raise rc.exceptions.CompletionException(self, str(e), "No weather information found")
    
    @classmethod
    def pretty_name(cls):
        return "Long Term Forecast"
```
